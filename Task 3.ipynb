```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/datasets/imtkaggleteam/netflix/NetFlix.csv')
df.head()
```

```python
df.info()
```

```python
sns.countplot(x='type', data=df)
plt.title('Movies vs TV Shows')
plt.show()
```

Observation: Movies are more prevalent than TV Shows on Netflix.

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Countries by Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.show()
```

Observation: The United States contributes the highest amount of Netflix content.

```python
df['date_added'] = pd.to_datetime(df['date_added'])

df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(10,5))

sns.countplot(x='year_added', data=df)

plt.xticks(rotation=90)

plt.title('Content Added Over Years')

plt.show()
```

Observation: Netflix content additions increased rapidly after 2016.

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)

plt.title('Distribution of Ratings')

plt.show()
```

Observation: TV-MA is one of the most common ratings.

```python
plt.figure(figsize=(10,6))

df['genres'].value_counts().head(10).plot(kind='barh')

plt.title('Top 10 Genres')

plt.show()
```

Observation: Some genres dominate the Netflix catalog.

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Movies vs TV Shows Percentage')

plt.ylabel('')

plt.show()
```

Observation: Movies account for the majority of Netflix content.

**Conclusion**

1. Movies dominate the Netflix platform.
2. The USA contributes the largest share of content.
3. Netflix content growth accelerated after 2016.
4. Certain ratings and genres are significantly more common.
5. Data visualization helps uncover patterns and trends effectively.
